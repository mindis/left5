```python
import numpy as np
import scipy as cp
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
```

```python
%matplotlib inline
```

```python
da = pd.read_csv('data/Chapter 1/m-gm3dx7508.txt', sep='\s+')
```

```python
gm = da['gm']
```

```python
plt.figure(figsize=(16, 4))
plt.plot(gm)
```

```python
acf_gm = acf(gm, nlags=24)
```

```python
plt.figure(figsize=(16, 4))
ind = np.arange(acf_gm.shape[0])
plt.bar(ind, acf_gm, width=0.1)
```
