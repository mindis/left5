```python
import os
import commands
import string
import shutil
import numpy as np
```

```python
def ext_rar(src_path, des_path):
    if not os.path.exists(des_path):
        os.mkdir(des_path)
    
    _msg = commands.getstatusoutput('rar x %s %s' % (src_path, des_path)) # (state_code, info) # last line: All OK
    if _msg[0] == 0:
        return True
    
    # todo log info _msg
    return False
```

```python
src_root = '/mnt/disk/weather2016'
dst_root = '/mnt/disk/weather_project/data_orig'

for sec_dir in os.listdir(src_root):
    sec_dir_root = os.path.join(src_root, sec_dir)
    
    for filename in os.listdir(sec_dir_root):
        filepath = os.path.join(sec_dir_root, filename)
        if os.path.isfile(filepath):
            name = filename.split('.', 1)[0]
            
            print 'Start extract: %s' % filepath
            r = ext_rar(filepath, os.path.join(dst_root, name))
            print 'End %s' % r
```

```python
def read_meta(meta_path='feature_meta2'):

    metas = []
    _first_path = ''
    for line in open(meta_path):
        _meta = map(string.strip, line.split('\t'))
        if len(_meta) > 1 and _meta[0]:
            _first_path = _meta[0]
        
        _meta[0] = _meta[0] if _meta[0] else _first_path
        if len(_meta) > 1 and not _meta[1]:
            continue
        metas.append(_meta)
    
    return metas
```

```python
meta = read_meta()
```

```python
def filter_data(meta, orig_data_path, data_path, date):
    print orig_data_path
    for _m in meta:
        _src_dir = os.path.join(orig_data_path, *_m)
        if not os.path.exists(_src_dir): continue
        _dst_dir = os.path.join(data_path, date, '-'.join(_m))
        if not os.path.exists(_dst_dir): os.makedirs(_dst_dir)
        
#         print _src_dir, _dst_dir
        
        _default_file = '%s08.000' % date
        cp_files_end = ['08.000', '08.003', '08.006', '08.009', '20.000', '20.003', '20.006', '20.009']
        file_names = os.listdir(_src_dir)
        file_names.sort()
        
        if _default_file not in file_names:
            _default_file = file_names[0]
            
        for _end in cp_files_end:
            cp_file = '%s%s' % (date, _end)
            if cp_file not in file_names:
                cp_file = _default_file
            
            src, dst = os.path.join(_src_dir, cp_file), os.path.join(_dst_dir, '%s%s' % (date, _end))
#             print 'cp file: %s -> %s' % (src, dst)
            shutil.copy2(src, dst)
            
            _default_file = cp_file
```
