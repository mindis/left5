```python
import os
import numpy as np
import torch
import matplotlib.pyplot as plt
import import_ipynb
from torch.utils.data import DataLoader
from torchvision import transforms, datasets

import torch.utils.data as data
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
```

```python
from datasets.jinnan import JinnanDetection
```

```python
def collate(batch):
    data = [item[0] for item in batch]
    target = [item[1] for item in batch]
    return [data, target]
```

```python
def collate_fn(data):
    data.sort(key=lambda x: len(x[1]), reverse=True)
    images, captions = zip(*data)

    images = torch.stack(images, 0)
     
    return images, captions
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
])

# transform = transforms.Compose([
#     transforms.Scale(256),
#     transforms.RandomCrop(224),
#     transforms.ToTensor()
# ])
```

```python
batch_size = 1
dataset = JinnanDetection('../data/First_round_data/jinnan2_round1_train_20190305', 'train_no_poly.json', transform=transform)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=1, collate_fn=collate_fn)
```

```python
fig = plt.figure()

for i, (imgs, labels) in enumerate(dataloader):
    print(imgs.shape)
#     print(imgs)
    print(labels)
    
    ax = fig.add_subplot(1, 1, i+1)
    ax.imshow(imgs[0].numpy().transpose([1,2,0]))
    
    break
```
